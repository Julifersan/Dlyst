```python
# S1

# Exercise_1: Mostrar las primeras filas del dataframe(entidad)
entidad.head()

```

```python
# S2

# Calcular el total de ventas por regiﾃｳn

# Ventasporegion : VAriable que almacena la operaciﾃｳn
# .groupby() : Funciﾃｳn que permite agrupar por un parﾃ｡metro, en este caso, el parﾃ｡metro serﾃ｡ un atributo de la entidad
  # El parﾃ｡metro se deberﾃ｡ pasar entre comillas, sean simples o dobles.
# .groupby("mientidad")["atributo_de_mientidad"]
# .sum() : Funciona sumando los valores de cada registro del subnivel.
# Subnivel : Para este caso, el subnivel de la regiﾃｳn serﾃｭa: ["Ventas"]
Ventasporregion = entidad.groupby("Region")["Ventas"].sum()


# Se muestran los resultados 筮ｸ十nprint(Ventasporregion)

```

```python
# S3

# Filtrar el dataframe para mostra solo las ventas realizads por "David"

# ventasDavid : Es el objeto o variable contenedora
# entidad[entidad["vendedor"]] == "David" : se toma de la entidad, un atributo de la misma, en la cual, se especifica el subnivel que queremos realizar el proceso
  # Es un poco engorroso la sintaxis, propondrﾃｭa esto : Entidad["atributo"]== "subnivel"
  # Puesto que se presta para malentendidos, o sere yo el que esta liando las cosas?
  # Es cuestiﾃｳn de costumbre supongo.

ventasDavid = entidad[entidad["Vendedor"] == "David"]

# Se muestran los resultados 筮ｸ十nprint(ventasDavid)
```

```python
# S4

# Calcular el promedio de ventas por regiﾃｳn

# Serﾃｭa similar al segundo ejercﾃｭcio, solo que en lugar de poner .sum(), se usa .mean()

Promedioventasregion = entidad.groupby("Region")["Ventas"].mean()

# Se muestra el resultado 筮ｸ十nprint(Promedioventasregion)
```

```python
# S5

# Calcular el promedio de ventas por region con dos decimales

# Lo mismo que el anterior, pero esta vez queremos el resultado con dos decimales
  # Para ello, se usarﾃ｡ round()
    # round(valor, cantidad de decimales)

# Pvr = Promedio de ventas por regiﾃｳn
Pvr = entidad.groupby("Region")["Ventas"].mean()
Pvr = round(Pvr, 2)

# Se muestra el resultado 筮ｸ十nprint(Pvr)
```

```python
# S6

# Ordenar el dataframe por ventas en orden descendente

# .sort_value()): Funcion que ordena de los valores iterables en forma ascendente de manera predeterminada, sin embargo, es buena prﾃ｡ctica estipularlo.
  # ascending = : True, para valores de forma ascendente (viene predeterminado)
  # ascending = False, para valores de forma descendente
  # .sort_value("atributo", ascending = False)
  # Esta funciﾃｳn es especﾃｭfica de pandas, aunque hace lo mismo que su homologo sorted() que viene integrada nativamente en python.


# Ovd : Orden ventas descendente
Ovd = entidad.sort_values("Ventas")
Ovd1 = entidad.sort_values("Ventas", ascending=True)

# Ambos imprimiran lo mismo, para que sea diferente (descendente), se debe usar ascending = False
# Se muestra el resultado 筮ｸ十nprint(Ovd)
print("\n\n筮ｸ十
\n")
print(Ovd1)
```

```python
# S7

# Filtrar las filas con ventas mayores a 1000 y ordenar de mayor a menor

# entidad[entidad["ventas"] > 1000] : De la entidad, selecciona la entidad y de estﾃ｡ el atributo de "Ventas", en donde este muestre los valores mayores a 1000
  # No me gusta como es la sintaxis, puesto que en pseudocﾃｳdigo, pensando como escribirlo, me explicarﾃｭa tﾃ｡l cual como lo escribi
  # Deberﾃｭa : De la entidad, toma el "atributo" y solo muestrame los valores mayores a 1000 : entidad["valores"] > 10000
filtrado = entidad[entidad["Ventas"] > 10000]

# Despues de seleccionado los datos que queremos, tomamos la variable y le aplicamos un orden para verlo cﾃｳmo nos pide el enunciado
ordenado = filtrado.sort_values("Ventas", ascending=False)

# Se muestran los resultados filtrados 筮ｸ十nprint(filtrado)

print("\n\n筮ｸ十
\n")

# Se muestran los resultados filtrados y ordenados 筮ｸ十nprint(ordenado)
```

```python
# S8

# Ordenar el dataframe por ventas en orden descendente

# Ovda = Orden ventas descendente again

Ovda = entidad.sort_values("Ventas", ascending =False)

# Se muestra el resultado 筮ｸ十nprint(Ovda)
```

```python
# S9

# Eliminar la columna "Vendedor"

# .drop(columns="Columna") : Elimina una columna especifica
# drop(columns=['Ventas', 'Productos']) : Elimina la columna "ventas" y "Productos"
# drop(index=[nﾃｺmero]) : Elimina una fila especifica, dada por un indice.
# drop(index=[0, 2]) : Elimina filas con indice 0 y 2

# Eseller = Eliminar Vendedor
Eseller = entidad.drop(columns="Vendedor")

# Se muestra el resultado 筮ｸ十nprint(Eseller)
```

```python
# S10

# Calcular el total de ventas por regiﾃｳn y producto

# Agrupame la entidad por el atributo regiﾃｳn, y suma las Ventas en cada regiﾃｳn
# Vpr = Ventas por regiﾃｳn
Vpr = entidad.groupby("Region")["Ventas"].sum()

# Agrupame la entidad por regiﾃｳn y suma las ventas por producto
# Vpp = Ventas por producto
Vpp = entidad.groupby("Producto")["Ventas"].sum()

# Muestrame el resultado 筮ｸ十nprint(Vpr)

print("\n\n        筮ｸ十
\n")

# Muestrame el resultado 筮ｸ十nprint(Vpp)
```
