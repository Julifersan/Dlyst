```python
import pandas as pd
import sqlite3
```

```python
# Cargar el archivo CSV en un DataFrame de Pandas
ruta = r'D:\01 Julifersan\02 Personal_Brand\02 Github\Dlyst\TalentoTech\Proyecto\Subsidios de vivienda asignados\fact_Subsidios.csv'

data = pd.read_csv(ruta)

# Crear una conexión a una base de datos SQLite en memoria
conex = sqlite3.connect(':memory:')

# Escribir el DataFrame en la base de datos SQLite
data.to_sql('tabla', conex)


query = """ select *
            from tabla;
        """

consulta = pd.read_sql(query, conex)

# Mostrar el resultado
consulta

```

# Cúal es el valor total asignado para el departamento de cundinamarca, en los años, 2008, 2016 y 2024?

```python
# Evitar notación científica
pd.set_option('display.float_format', '{:.0f}'.format)

query1 = """Select Departamento, Año_Asignación, SUM(Valor_Asignado) as 'Valor Total Asignado'
            From tabla
            where Departamento = 'CUNDINAMARCA' AND Año_Asignación IN (2008, 2016, 2024)
            group by Departamento, Año_Asignación
"""
Uno = pd.read_sql(query1, conex)

Uno
```

# Cúal es el promedio del valor asignado de los subsidios en los años 2020, 2012, 2005?

```python
query2 = """select Año_Asignación, AVG(Valor_Asignado) as 'Promedio Valor Asignado'
            from tabla
            where Año_Asignación IN (2020, 2012, 2005)
            group by Año_Asignación;
"""

Dos = pd.read_sql(query2, conex)

Dos
```

# Cúal es el valor asignado para el año con menor presupuesto?

```python
query3 = """select Año_Asignación, Valor_Asignado as 'Sumatoria Valor Asignado'
            from tabla
            group by Año_Asignación
            order by Valor_Asignado ASC
            Limit 1
"""

Tres = pd.read_sql(query3, conex)

Tres
```

# Cúal es el valor asignado para los 5 departamento con mayor presupuesto?

```python
# Mostrar con formato de números grandes
pd.set_option('display.float_format', '{:,.0f}'.format)

query4 = """select Departamento, Valor_Asignado
            from tabla
            group by Departamento
            order by SUM(Valor_Asignado) DESC
            LIMIT 5
"""

Cuatro = pd.read_sql(query4, conex)

Cuatro
```

# Cúal es el valor asignado para los 5 departamento con menor presupuesto?

```python
query5 = """select Departamento, SUM(Valor_Asignado) as Valor_Total_Asignado
            from tabla
            group by Departamento
            order by SUM(Valor_Asignado) ASC
            LIMIT 5;
"""
Cinco = pd.read_sql(query5, conex)
Cinco
```

# Cúales son los 4 años con menor presupuesto?

```python
query6 = """select Año_Asignación, SUM(Valor_Asignado)
            from tabla
            group by Año_Asignación
            ORDER BY Valor_Asignado ASC
            LIMIT 4;
"""
Seis = pd.read_sql(query6, conex)

Seis
```
