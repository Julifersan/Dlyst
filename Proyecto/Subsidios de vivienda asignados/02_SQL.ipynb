```python
import pandas as pd
import sqlite3

# Cargar el archivo CSV en un DataFrame de Pandas
ruta =r'cleanData.csv'

data = pd.read_csv(ruta)

# Crear una conexión a una base de datos SQLite en memoria
conex = sqlite3.connect(':memory:')

# Escribir el DataFrame en la base de datos SQLite
data.to_sql('tabla', conex)


query = """select *
                    from tabla;"""

consulta = pd.read_sql(query, conex)

# Mostrar el resultado
consulta

```

**1. Tipo de licencias únicas**

```python
query1 = """select 
                        distinct(TIPO_DE_LICENCIA)
                    from tabla;"""

consultTipo_licencia = pd.read_sql(query1, conex)

consultTipo_licencia
```

**2. Cuántos tipos de _estados_ existen en la data**

```python
query2 = """select distinct(estado)
                    from tabla;
"""
consultaEstado2 = pd.read_sql(query2, conex)

consultaEstado2
```

**3. Cúal departamento tiene el mayor número de licencias vigentes**

```python
query3 = """select 
                        ciudad, 
                        depto, 
                        count(Estado) as Cantidad_LicenciasVigentes
                    from 
                        tabla
                    WHERE 
                        Estado = 'LICENCIA VIGENTE'
                    group by 
                        depto
                    order by 
                        Cantidad_LicenciasVigentes desc;
                    """

consultaEstado3 = pd.read_sql(query3, conex)

consultaEstado3
```

**4. Cantidad de LICENCIAS EXTENDIAS por departamento**

```python
query4 = """select ciudad, departamento, count(Estado) as cantidadLicenciasExtendidas
                    from tabla
                    where Estado = 'LICENCIA EXTENDIDA'
                    group by departamento
                    order by cantidadLicenciasExtendidas desc;
                    """

consulta4 = pd.read_sql(query4, conex)

consulta4
```

**5. Cantidad de LICENCIA CANCELADA por departamento**

```python
query5 = """select ciudad, departamento, count(Estado) as CantidadLicenciaCancelada
                    from tabla
                    where Estado = 'LICENCIA CANCELADA'
                    group by departamento
                    order by CantidadLicenciaCancelada desc;
"""

consulta5 = pd.read_sql(query5, conex)

consulta5
```

**6. Cantidad de LICENCIA VENCIDAD por departamento**

```python
query6 = """select ciudad, departamento, count(Estado) as CantidadLicenciaVencida
                    from tabla
                    where Estado = 'LICENCIA VENCIDA'
                    group by departamento
                    order by CantidadLicenciaVencida desc;
"""
consulta6 = pd.read_sql(query6, conex)

consulta6
```

**7. Años evaluados**

```python
query7 = """
    SELECT distinct( strftime('%Y', Fecha)) as año
    FROM tabla
    order by año asc;
"""

consulta7 = pd.read_sql(query7, conex)

consulta7
```

**8. Año en el que más se han dado LICENCIA VIGENTE**

```python
query8 ="""
                    select strftime('%Y', Fecha) as Año, count(Estado) as CantidadLicenciasVigente
                    from tabla
                    where Estado = 'LICENCIA VIGENTE';

"""

consulta8 = pd.read_sql(query8, conex)

consulta8
```

**9. 4 primeros meses con su año que más se han dado LICENCIA VIGENTE**

```python
query9 = """
                    select  strftime('%Y', Fecha) as Año, 
                                strftime('%m', Fecha) as Mes, 
                                count(*) as CantidadLicenciaVigente
                    from tabla
                    where Estado = 'LICENCIA VIGENTE'
                    group by Año, Mes
                    order by CantidadLicenciaVigente desc
                    limit 4;
"""

consulta9 = pd.read_sql(query9, conex)

consulta9
```
