```python
import pandas as pd
```

```python
archivo = r"/content/drive/MyDrive/Colab Notebooks/Data/Dataframe2.csv"

data2 = pd.read_csv(archivo, sep=";")

print(data2)
```

```python
r_data2 = data2.rename(columns={
    "id":"id",
    "weight":"PESO",
    "height":"ALTURA",
    "age":"EDAD"
    })

print(r_data2)
```

```python
converpeso = 0.453592
converaltura = 0.3048

cpdata2 = r_data2["PESO"] * converpeso
cadata2 = r_data2["ALTURA"] * converaltura

newdata2 = pd.DataFrame({
    "id":r_data2["id"],
    "PESO":cpdata2,
    "ALTURA":cadata2,
    "EDAD": r_data2["EDAD"]
})

print(newdata2)

```
