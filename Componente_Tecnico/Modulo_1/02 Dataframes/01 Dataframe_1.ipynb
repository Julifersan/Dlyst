```python
import pandas as pd
```

```python
# Se establece la ruta del archivo y se con ello se cra un dataframe llamada data1
archivo = r"Dataframe1.csv"

data1 = pd.read_csv(archivo, sep=";")

# Se muesttra el resultado ⬇️
data1
```

```python
# Se crea una nuevo contenedor, el cual tendra todos los atributos de la entidad con los nombres en español
renombrada_data1 = data1.rename(columns = {
    "weight" : "PESO", 
    "height":  "ALTURA", 
    "age":     "EDAD", 
    "gender":  "GENERO" 
    })

# Se muestra el resultado ⬇️

renombrada_data1
```
