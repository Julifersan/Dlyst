```python
import pandas as pd
```

```python
# Se establece la ruta del archivo y se con ello se cra un dataframe llamada data1
archivo = r"/content/drive/MyDrive/Colab Notebooks/Data/Dataframe1.csv"

data1 = pd.read_csv(archivo, sep=";")

# Se muesttra el resultado ⬇️
print(data1)
```

```python
# Se crea una nuevo contenedor, el cual tendra todos los atributos de la entidad con los nombres en español
renombrada_data1 = data1.rename(columns = {"weight" : "PESO", "height":"ALTURA", "age":"EDAD", "gender":"GENERO" })

# Se muestra el resultado ⬇️
print(renombrada_data1)
```

```python
# Se hace una conversión de valores

# Factores constantes de conversión
convertpeso = lbs_to_Kg = 0.453592
convertaltura = pies_to_mt = 0.3048

# Se realizan las conversiones
Data_conversion_peso = renombrada_data1["PESO"] * convertpeso
Data_conversion_altura = renombrada_data1["ALTURA"] * convertaltura

Data_conversion_peso = round(Data_conversion_peso, 2)
Data_conversion_altura = round(Data_conversion_altura, 2)

# Se crea un nuevo dataframe con los valores convertidos
newdata = pd.DataFrame({
    "id":renombrada_data1['id'],
    "PESO": Data_conversion_peso,
    "ALTURA": Data_conversion_altura,
    "EDAD": renombrada_data1["EDAD"],
    "GENERO": renombrada_data1["GENERO"]
    })

# Se muestra el resultado ⬇️
print(newdata)
```
