```python
# S1

# Exercise_1: Mostrar las primeras filas del dataframe(entidad)
entidad.head()

```

```python
# S2

# Calcular el total de ventas por región

# Ventasporegion : VAriable que almacena la operación
# .groupby() : Función que permite agrupar por un parámetro, en este caso, el parámetro será un atributo de la entidad
  # El parámetro se deberá pasar entre comillas, sean simples o dobles.
# .groupby("mientidad")["atributo_de_mientidad"]
# .sum() : Funciona sumando los valores de cada registro del subnivel.
# Subnivel : Para este caso, el subnivel de la región sería: ["Ventas"]
Ventasporregion = entidad.groupby("Region")["Ventas"].sum()


# Se muestran los resultados ⬇️
print(Ventasporregion)

```

```python
# S3

# Filtrar el dataframe para mostra solo las ventas realizads por "David"

# ventasDavid : Es el objeto o variable contenedora
# entidad[entidad["vendedor"]] == "David" : se toma de la entidad, un atributo de la misma, en la cual, se especifica el subnivel que queremos realizar el proceso
  # Es un poco engorroso la sintaxis, propondría esto : Entidad["atributo"]== "subnivel"
  # Puesto que se presta para malentendidos, o sere yo el que esta liando las cosas?
  # Es cuestión de costumbre supongo.

ventasDavid = entidad[entidad["Vendedor"] == "David"]

# Se muestran los resultados ⬇️
print(ventasDavid)
```

```python
# S4

# Calcular el promedio de ventas por región

# Sería similar al segundo ejercício, solo que en lugar de poner .sum(), se usa .mean()

Promedioventasregion = entidad.groupby("Region")["Ventas"].mean()

# Se muestra el resultado ⬇️
print(Promedioventasregion)
```

```python
# S5

# Calcular el promedio de ventas por region con dos decimales

# Lo mismo que el anterior, pero esta vez queremos el resultado con dos decimales
  # Para ello, se usará round()
    # round(valor, cantidad de decimales)

# Pvr = Promedio de ventas por región
Pvr = entidad.groupby("Region")["Ventas"].mean()
Pvr = round(Pvr, 2)

# Se muestra el resultado ⬇️
print(Pvr)
```

```python
# S6

# Ordenar el dataframe por ventas en orden descendente

# .sort_value()): Funcion que ordena de los valores iterables en forma ascendente de manera predeterminada, sin embargo, es buena práctica estipularlo.
  # ascending = : True, para valores de forma ascendente (viene predeterminado)
  # ascending = False, para valores de forma descendente
  # .sort_value("atributo", ascending = False)
  # Esta función es específica de pandas, aunque hace lo mismo que su homologo sorted() que viene integrada nativamente en python.


# Ovd : Orden ventas descendente
Ovd = entidad.sort_values("Ventas")
Ovd1 = entidad.sort_values("Ventas", ascending=True)

# Ambos imprimiran lo mismo, para que sea diferente (descendente), se debe usar ascending = False
# Se muestra el resultado ⬇️
print(Ovd)
print("\n\n⬇️\n\n")
print(Ovd1)
```

```python
# S7

# Filtrar las filas con ventas mayores a 1000 y ordenar de mayor a menor

# entidad[entidad["ventas"] > 1000] : De la entidad, selecciona la entidad y de está el atributo de "Ventas", en donde este muestre los valores mayores a 1000
  # No me gusta como es la sintaxis, puesto que en pseudocódigo, pensando como escribirlo, me explicaría tál cual como lo escribi
  # Debería : De la entidad, toma el "atributo" y solo muestrame los valores mayores a 1000 : entidad["valores"] > 10000
filtrado = entidad[entidad["Ventas"] > 10000]

# Despues de seleccionado los datos que queremos, tomamos la variable y le aplicamos un orden para verlo cómo nos pide el enunciado
ordenado = filtrado.sort_values("Ventas", ascending=False)

# Se muestran los resultados filtrados ⬇️
print(filtrado)

print("\n\n⬇️\n\n")

# Se muestran los resultados filtrados y ordenados ⬇️
print(ordenado)
```

```python
# S8

# Ordenar el dataframe por ventas en orden descendente

# Ovda = Orden ventas descendente again

Ovda = entidad.sort_values("Ventas", ascending =False)

# Se muestra el resultado ⬇️
print(Ovda)
```

```python
# S9

# Eliminar la columna "Vendedor"

# .drop(columns="Columna") : Elimina una columna especifica
# drop(columns=['Ventas', 'Productos']) : Elimina la columna "ventas" y "Productos"
# drop(index=[número]) : Elimina una fila especifica, dada por un indice.
# drop(index=[0, 2]) : Elimina filas con indice 0 y 2

# Eseller = Eliminar Vendedor
Eseller = entidad.drop(columns="Vendedor")

# Se muestra el resultado ⬇️
print(Eseller)
```

```python
# S10

# Calcular el total de ventas por región y producto

# Agrupame la entidad por el atributo región, y suma las Ventas en cada región
# Vpr = Ventas por región
Vpr = entidad.groupby("Region")["Ventas"].sum()

# Agrupame la entidad por región y suma las ventas por producto
# Vpp = Ventas por producto
Vpp = entidad.groupby("Producto")["Ventas"].sum()

# Muestrame el resultado ⬇️
print(Vpr)

print("\n\n        ⬇️\n\n")

# Muestrame el resultado ⬇️
print(Vpp)
```
